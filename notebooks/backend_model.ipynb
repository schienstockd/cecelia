# Cecelia — Backend Model

```julia
import Pkg; Pkg.activate(joinpath(@__DIR__, "..", "app"))

include("../app/src/Cecelia.jl")
using .Cecelia

init_cecelia!()   # reads ~/cecelia-pineapple/dev/config.toml
```

## Create project

```julia
proj = create_project!(name="Live Cell Imaging", kind="live")
```

## Add experimental sets and images

```julia
ctrl    = add_set!(proj, name="Control")
treated = add_set!(proj, name="Treated")
img     = add_image!(ctrl, name="M1-1 z300 t004")
```

## Import raw file as OME-ZARR

```julia
raw = expanduser("~/Downloads/M1-1-B6-naive-gBT-uGFP-OTI-CTV-P14-ubTomato-z300_0004-1(1).tif")
run_task(ImportOmezarr(), img, Dict{String,Any}("src_path" => raw))
```

## Add metadata and save

```julia
img.meta["channel_names"] = ["gBT uGFP", "OTI CTV", "P14 ubTomato", "Ch4"]
save!(proj)   # cascades: project → sets → images
```
